### IMDB データセットの単語数のヒストグラム

参考： https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
import tensorflow_datasets as tfds

data = tfds.load('imdb_reviews')
data_train = data['train']
tokenizer = tfds.deprecated.text.Tokenizer()
list_len = []
for v in data_train:
    tokenized = tokenizer.tokenize(v['text'].numpy())
    list_len.append(len(tokenized))
```

```python
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline

plt.figure(figsize=(10, 4))
p = sns.histplot(list_len)
p.set_xlabel('length of movie reviews')
plt.show()

plt.figure(figsize=(10, 4))
p = sns.histplot(list_len, log_scale=(False, True))
p.set_xlabel('length of movie reviews')
plt.show()
```
